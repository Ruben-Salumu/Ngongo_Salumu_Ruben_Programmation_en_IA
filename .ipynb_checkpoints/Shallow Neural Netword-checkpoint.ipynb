```python
sub_data = data[['pktcount','bytecount','Protocol','src','pktrate','pktperflow','dur_nsec','switch','dst','label']]
```

# Diviser le train_data & test_data

```python
X = sub_data.iloc[:, :-1] #ici on prend toute les lignes et toutes les colonnes sauf la colonne label
y = sub_data.iloc[:, -1] #ici on prend toute les lignes et la derniere colonne la colonne label
```

```python
X
```

```python
y
```

```python
# creation de X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.3, random_state=35)
```

```python
print("X_train Shape", X_train.shape)
print("y_train Shape", y_train.shape)
print("X_test Shape", X_test.shape)
print("y_test Shape", y_test.shape)
```

```python
sns.heatmap(sub_data.corr(),annot=True)
```

```python
# Standarisation
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

# 6 Construire le Shallow Neural Netword

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
#Creer une instance de la classe sequencial
model = Sequential()
```

```python
#Ajout de la premier couche
model.add(Dense(10, input_shape=(9,), activation='relu'))

#Ajout de la couche de sortie
model.add(Dense(1, activation='sigmoid'))

#compiler le model
model.compile(optimizer=Adam(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])

#afficher de l'architecture du reseau de neuronne
model.summary()
```

```python
#Entrainement du model
#model.fit(X_train, y_train, epochs=10, verbose=1)
```

```python
#model.fit(X_train, y_train, epochs=10, verbose=1, validation_data=(X_test, y_test))
```

```python
history = model.fit(X_train, y_train, epochs=50, verbose=1, validation_data=(X_test, y_test))
```

```python
#7 Visualiser l'évolution de l'entrainement du reseau de neurone

epoch_accuracy = history.history['accuracy']
epoch_loss = history.history['loss']

epoch_val_accuracy = history.history['val_accuracy']
epoch_val_loss = history.history['val_loss']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6, 4))
plt.plot(range(0, len(epoch_accuracy)), epoch_accuracy, color='blue', label='Training Accuracy')
plt.plot(range(0, len(epoch_loss)), epoch_loss, color='red', label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('values')
plt.title('Accuracy vs Loss')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(range(0, len(epoch_accuracy)), epoch_accuracy, color='blue', label='Training Accuracy')
plt.plot(range(0, len(epoch_loss)), epoch_val_accuracy, color='red', label='Test Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training Accuracy vs Test Accuracy')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(range(0, len(epoch_loss)), epoch_loss, color='blue', label='Training Loss')
plt.plot(range(0, len(epoch_val_loss)), epoch_val_loss, color='red', label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Accuracy Vs Taining Loss')
plt.legend()
plt.show()
```
